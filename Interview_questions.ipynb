```python
def create_board():
    """Creates an empty Connect 4 board."""
    return [[' ' for _ in range(7)] for _ in range(6)]

def print_board(board):
    """Prints the current state of the board."""
    for row in reversed(board):
        print('|', end='')
        for cell in row:
            print(f' {cell} |', end='')
        print()
    print('+' + '---+' * 7)
    print('  1   2   3   4   5   6   7  ')

def drop_piece(board, col, piece):
    """Drops a piece into the specified column."""
    for row in range(6):
        if board[row][col] == ' ':
            board[row][col] = piece
            return True
    return False

def check_win(board, piece):
    """Checks if the given piece has won."""
    # Check horizontal
    for row in range(6):
        for col in range(4):
            if all(board[row][col + i] == piece for i in range(4)):
                return True

    # Check vertical
    for row in range(3):
        for col in range(7):
            if all(board[row + i][col] == piece for i in range(4)):
                return True

    # Check diagonals (positive slope)
    for row in range(3):
        for col in range(4):
            if all(board[row + i][col + i] == piece for i in range(4)):
                return True

    # Check diagonals (negative slope)
    for row in range(3):
        for col in range(3, 7):
            if all(board[row + i][col - i] == piece for i in range(4)):
                return True

    return False

def play_game():
    """Simulates a Connect 4 game."""
    board = create_board()
    current_player = 'X'

    while True:
        print_board(board)
        col = int(input(f"Player {current_player}, choose a column (1-7): ")) - 1

        if not 0 <= col < 7:
            print("Invalid column. Please choose a column between 1 and 7.")
            continue

        if not drop_piece(board, col, current_player):
            print("Column is full. Please choose another column.")
            continue

        if check_win(board, current_player):
            print_board(board)
            print(f"Player {current_player} wins!")
            break

        if all(cell != ' ' for row in board for cell in row):
            print_board(board)
            print("It's a tie!")
            break

        current_player = 'O' if current_player == 'X' else 'X'

if __name__ == "__main__":
    play_game()
```

```python
import heapq
from collections import deque
class Word_sequence:
  def __init__(self,word_items):
    self.word_items = word_items
    print(self.word_items)

  def greedy_search(self,first_word):
    results_words = [first_word]
    while first_word:
      current_word = first_word
      max_val = 0
      max_key = ""
      for k,v in self.word_items.get(current_word,{}).items():
          max_val = max(max_val,v)
          if v == max_val:
            max_key = k
      first_word = max_key if max_key != "." else None
      results_words.append(max_key)
    print(results_words)

  def greedy_search1(self,first_word):
    results_words = [first_word]
    while first_word:
      current_word = first_word
      max_val = 0
      max_key = ""
      word_map = self.word_items.get(current_word,{})
      word_heap = [(-v,k) for k,v in word_map.items()]
      print(word_heap)
      heapq.heapify(word_heap)
      print(word_heap)
      max_key = heapq.heappop(word_heap)[1]
      first_word = max_key if max_key != "." else None
      results_words.append(max_key)
    print(results_words)

  def beam_search(self, start_word="the", beam_width=3):
      if start_word not in self.word_items:
          return 0, []

      beams = deque([(start_word, 0, [start_word])])  # (word, score, sequence)

      while beams:
          new_beams = []
          for _ in range(len(beams)):
              current_word, current_score, current_sequence = beams.popleft()

              if current_word == ".":
                  new_beams.append((current_word, current_score, current_sequence))
                  continue

              next_words = self.word_items.get(current_word, {})
              for next_word, score in next_words.items():
                  new_beams.append((next_word, current_score + score, current_sequence + [next_word]))

          # Keep only the top beam_width beams based on score
          new_beams.sort(key=lambda x: x[1], reverse=True)
          beams = deque(new_beams[:beam_width])

      # Find the beam with the highest score
      best_beam = max(beams, key=lambda x: x[1])
      return best_beam[1], best_beam[2]

ws = {
  "the" :{
  "the" : 2,
  "quick":3,
  "brown" : 4,
  "fox": 6,
  "jumped" : 1,
  "building" : 8,
  "." : 0

  },
  "quick" : {
  "the" : 2,
  "quick":3,
  "brown" : 4,
  "fox": 3,
  "jumped" : 1,
  "building" : 1,
  "." : 0
  },
  "brown" : {
    "the" : 2,
  "quick":3,
  "brown" : 0,
  "fox": 10,
  "jumped" : 1,
  "building" : 8,
  "." : 0
  }
  ,
  "fox" : {
    "the" : 2,
  "quick":3,
  "brown" : 4,
  "fox": 0,
  "jumped" : 15,
  "building" : 5,
  "." : 0
  },

  "jumped" :
  {"the" : 12,
  "quick":3,
  "brown" : 4,
  "fox": 6,
  "jumped" : 0,
  "building" : 8,
  "." : 0},

  "building" : {"the" : 2,
  "quick":3,
  "brown" : 4,
  "fox": 6,
  "jumped" : 1,
  "building" : 0,
  "." : 15},
  "." : {
    "." : 0,
    "." : 0,
    "." : 0,
    "." : 0,
    "." : 0,
    "." : 0,
  }

  }
Word_sequence(ws).greedy_search("the")
Word_sequence(ws).greedy_search1("the")
Word_sequence(ws).beam_search("the")
```
